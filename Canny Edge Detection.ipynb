### Canny Edge Detector

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Lendo a imagem
image = cv2.imread('images/brain_MR.jpg')

# Mudar cor de BGR para RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Convertendo a imagem para cinza para processamento
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

### Implementar o algoritmo Canny Edge Detection

```python
# Try Canny using "wide" and "tight" thresholds
# Applicando usando limiares largo e justo

wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 200, 240)
 
    
# Mostrar as imagens
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('wide')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight, cmap='gray')
```

```python
# Lendo a imagem
image = cv2.imread('images/sunflower.jpg')

# Mudar cor de BGR para RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Convertendo a imagem para escala de cinza
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)


# Defina limiares superior e inferior para histerese
# Para tais valores, estão tão pequenos e baixos que irá mostrar ruído
# tentar 100,200
lower = 0
upper = 20


edges = cv2.Canny(gray, lower, upper)

plt.figure(figsize=(20,10))
plt.imshow(edges, cmap='gray')
```
